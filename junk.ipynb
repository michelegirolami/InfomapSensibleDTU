```python
def community_detection_igraph(nodes,edges):
    """Return list of communities in network from igraph's implementation if Infomap
    
    Parameters
    ----------
    nodes : list
        Nodes, with unique names, e.g. hashed values
    edges : list of 3-tuples
        Weighted edges in network. If not weighted, provide equal
        weights for every edge.
        
    Returns
    -------
    communities_hash : list of lists
        Communities in the network, inferred with the map equation in igraph.
    """
    g = Graph()
    g.add_vertices(nodes)
    
    int_to_hash = dict(zip(range(len(nodes)),nodes))

    for v1,v2,w in edges:
        g.add_edge(v1, v2, weight=w)
        
    communities_int = list(g.community_infomap())
    communities_hash = [[int_to_hash[n] for n in com] for com in communities_int]
        
    return communities_hash
```
